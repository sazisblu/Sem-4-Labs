`1.4 Error Handling` 

```python
# Error handling

a = -2
try:
    if a < 0:
        raise ValueError("Cannot find the squareroot of negative number")
    
    b = math.sqrt(a)
    print(b)
except ValueError as e:
    print(e)

# Output:
# Cannot find the squareroot of negative number

try:
    c = 10 / 0  # This will raise a ZeroDivisionError
    raise ValueError("Cannot divide by zero")  # This won't execute because of the above error
except ZeroDivisionError as e:
    print("Caught a ZeroDivisionError:", e)
except ValueError as e:
    print("Caught a ValueError:", e)
finally:
    print("The error is of another source or handled completely.")

# Output:
# Caught a ZeroDivisionError: division by zero
# The error is of another source or handled completely.
```

`1.6 Graphics Visualisation`

**1. np.linspace(start, stop, num)** 
- This is a NumPy function that returns a set of evenly spaced numbers over a specified interval.

- Parameters: <br>
    start: The starting value of the sequence.

    stop: The ending value of the sequence.

    num: (the third argument) The number of values to generate between start and stop (inclusive).

**2. plt.plot(x, y)**
- Purpose: Plots y versus x as lines and/or markers.
- Defaults:
    It uses a blue solid line ('-b') by default.
    You can customize it with arguments like plt.plot(x, y, 'ro--') for red circles with dashed lines.

**3. plt.show( )**
- Purpose: Displays the figure window containing the plot.
- What it does:
    Without plt.show(), your plot may not appear—especially in non-interactive environments (like scripts or some IDEs).
    It starts the GUI event loop that renders the plot on screen.

**4. plt.figure()**
- Purpose: Creates a new figure (a blank canvas) for plotting.
- Use case:
    Useful when you're plotting multiple plots in a script.
    Allows you to customize figure size, DPI, etc.

**5. plt.xlabel() and plt.ylabel()**
 - Purpose: Adds labels to the x-axis and y-axis.

**6. plt.xlim() and plt.ylim()**
- Purpose: Sets limits for the x-axis and y-axis.

**7. plt.legend()**
- Purpose: Displays a legend for the plot (helpful when multiple lines are plotted).
- Needs: You must provide a label in plt.plot().

**8.loc**
- This parameter specifies the location of the legend.
-  Common loc Values.You can use either or you can also use numbers (0–10) instead of strings, but strings are more readable.

| **`loc` Value**  | **Position**                    |
| ---------------- | ------------------------------- |
| `'best'`         | Let matplotlib decide (default) |
| `'upper right'`  | Top-right corner                |
| `'upper left'`   | Top-left corner                 |
| `'lower left'`   | Bottom-left corner              |
| `'lower right'`  | Bottom-right corner             |
| `'center'`       | Center of plot                  |
| `'center left'`  | Middle left side                |
| `'center right'` | Middle right side               |
| `'upper center'` | Top center                      |
| `'lower center'` | Bottom center                   |

```python
plt.figure(figsize = (10,6)) # Set the figure size
x = np.linspace(-5,5,100) # 100 points between -5 and 5
plt.plot(x, x**2, "ko", label = "quadratic") # plot the quadratic function
plt.plot(x, x**3, "r*", label = "cubic") # plot the cubic function
plt.title(f"Plot of Various Polynomials from {x[0]} to {x[-1]} ") # gives title to the figure
plt.xlabel("X axis") # labels the x axis 
plt.ylabel ("Y axis") # labels the y axis
plt.legend(loc = "lower right") # adds a legend to the plot
plt.xlim(-6.6)  
plt.ylim(-10,10)
plt.grid() # turns on or off the grid lines
plt.show () 
```

**Given the lists x = np.arange(11) and y = x2, create a 2 x 3 subplot where each subplot plots x
versus y using plot, scatter, bar, loglog, semilogx, and semilogy. Title and label each plot
appropriately. Use a grid, but a legend is not necessary here.**

```python

x = np.arange (11)
y= x**2
plt.figure(figsize = (14, 8))
plt.subplot(2, 3, 1)
plt.plot (x,y)
plt.title ("Plot")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.grid()
plt.subplot(2, 3, 2)
plt.scatter (x,y)
plt.title ("Scatter")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.grid()
plt.subplot(2, 3, 3)
plt.bar (x,y)
plt.title ("Bar")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.grid()
plt.subplot(2, 3, 4)
plt.loglog (x, y)
plt.title ("Loglog")
plt.xlabel ("X")
plt.ylabel("Y")  
plt.grid(which="both")
plt.subplot (2, 3, 5)
plt.semilogx (x, y)
plt.title ("Semilogx")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.grid(which="both")
plt.subplot (2, 3, 6)
plt.semilogy (x,y)
plt.title ("Semilogy")
plt.xlabel ("X")
plt.grid()
plt.tight_layout ()
plt.show () 


```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-poster")
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')  # Add 3D axes

# Sample 3D data
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot3D(x, y, z, 'purple')

plt.show()

```
